#Exploratory Visuals & Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fpath='/content/drive/MyDrive/StudentFolders16wk24wk/CodingDojo/01-Fundamentals/Week02/Data/sales_predictions_2023.csv'
df=pd.read_csv(fpath)
df.head()
```

##Histograms to view the distributions of numerical features in your dataset.

```python
ax=df['Item_Weight'].hist(bins='auto',edgecolor='black')
ax.tick_params(axis='x')
ax.ticklabel_format(style='plain')
ax.set(title='Item weight',xlabel='item weight',ylabel='count');
```

the distribution of items weights is almsot similar to each other
we can see that the highest count is for weight=7.5 and almsot 9.

```python
ax=df['Item_Visibility'].hist(bins='auto',edgecolor='black')
ax.tick_params(axis='x')
ax.ticklabel_format(style='plain')
ax.set(title='Item_Visibility',xlabel='Item_Visibility',ylabel='count');
```

the distribution is higher with a low item visibility.

```python
ax=df['Item_MRP'].hist(bins='auto',edgecolor='black')
ax.tick_params(axis='x')
ax.ticklabel_format(style='plain')
ax.set(title='Item_MRP',xlabel='Item_MRP',ylabel='count');
```

the distribution is alsmot equal with different values and the highest is for the item MRP=120 almsot.

```python
ax=df['Outlet_Establishment_Year'].hist(bins='auto',edgecolor='black')
ax.tick_params(axis='x')
ax.ticklabel_format(style='plain')
ax.set(title='Outlet_Establishment_Year',xlabel='Outlet_Establishment_Year',ylabel='count');
```

the distribution here is higher for year 1985 and 1999.

```python
ax=df['Item_Outlet_Sales'].hist(bins='auto',edgecolor='black')
ax.tick_params(axis='x')
ax.ticklabel_format(style='plain')
ax.set(title='Item_Outlet_Sales',xlabel='Item_Outlet_Sales',ylabel='count');
```

the distribution is higher in the begining when sales=0 and gets lower as sales get higher.

##Heatmap to view the correlation between features.

```python
corr=df.corr(numeric_only=True)
corr.round()
```

```python
fig, ax=plt.subplots(figsize=(20,20))
sns.heatmap(corr,cmap="coolwarm",annot=True,ax=ax);
```
