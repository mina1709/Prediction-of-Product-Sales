##Part 1) Load the data and make a datetime index.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url ="/content/drive/MyDrive/StudentFolders16wk24wk/CodingDojo/04-DataViz/Week13/Data/london_weather_MODIFIED - london_weather_MODIFIED.csv"
df = pd.read_csv(url)
df
```

1-Convert the date column to datetime dtype

```python
pd.to_datetime(df['date'])
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['date']
```

2-Set the date column as the index.

```python
df = df.set_index('date')
df
```

3-Filter out only the required data

```python
df = df.loc['2000':, ['precipitation', 'mean_temp', 'min_temp',
                      'max_temp', 'snow_depth']]

df
```

4-Impute any missing values.

```python
df.isna().sum()
```

```python
df['mean_temp'] = df['mean_temp'].interpolate()
df['max_temp'] = df['max_temp'].interpolate()
df['min_temp'] = df['min_temp'].interpolate()
```

```python
df.isna().sum()
```

```python
df['precipitation'] = df['precipitation'].fillna(0)
df['snow_depth'] = df['snow_depth'].fillna(0)
```

```python
df.isna().sum()
```

```python
from matplotlib import dates as mdates
```

Q1

```python
precip_cols = ['precipitation']
plot_df  = df.loc["2000":"2010",precip_cols]
plot_df
```

```python
ax = plot_df.plot(figsize=(12,3))
ax.set( title='Precipitation for 2000-2010',
      ylabel='Inches')

# customize major ticks
major_loc =  mdates.YearLocator(1)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)

ax.xaxis.set_minor_locator(minor_loc)


## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)
```

Q2

```python
temp_cols = ['mean_temp']
plot_df = df.loc['2000':'2020',temp_cols]
plot_df
```

```python
ax = plot_df.plot(figsize=(12,5))
ax.set(title='Average Temperature', ylabel='Degrees')

# customize major ticks
major_loc =  mdates.YearLocator(5)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.YearLocator(1)
minor_fmt = mdates.DateFormatter("%b")

ax.xaxis.set_minor_locator(minor_loc)
ax.xaxis.set_minor_formatter(minor_fmt)

## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)
```
